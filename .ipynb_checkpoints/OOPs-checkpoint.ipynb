```python
## create the design of the object --> create a class

class Car:
    name = "Honda City"
    category = "Sedan"
    manufacture_year = 2025


## the problem is that name. category etc is hardcoded in the class itself. Thats not desirable
```

```python
## manufacture some car

car1 = Car()  # creating an instance of a car 


print(type(car1))  # creating a custom datatype in python
```

```python
car1.name  # accesing the attribute of the object
```

```python
car1.category
```

```python
car1.manufacture_year
```

```python
class Car:
    ## atribute
    name = None  
    category = None
    manufacture_year = None

    # method
    def start():
        print("this car is now started!")

    def drive():
        print("this car is driving!")


car1 = Car()
car2 = Car()
car3 = Car()
```

```python
car1.name = "Hyundai Verna"
car2.name = "Honda Jazz"
car3.name = "Tata Nexon"
```

```python
car1.category
```

```python
class Vehicle:

    def __init__(self, make, model, year, fuel_type):
        print("Vehicle is being created!!")
        self.make      = make
        self.model     = model
        self.year      = year
        self.fuel_type = fuel_type
        self.speed     = 0   # all vehicles start at rest

    def accelerate(self, amount):
        self.speed += amount
        print(f"{self.make} {self.model} accelerating to {self.speed} km/h")

    def brake(self, amount):
        self.speed = max(0, self.speed - amount)
        print(f"{self.make} {self.model} slowing to {self.speed} km/h")

    def __str__(self): ## how the object is serialised --> for when the object is to be shown in string or printed
        return f"{self.year} - {self.make} - {self.model} - ({self.fuel_type})"

    def __repr__(self): ## how the object is represented
        return f"{self.year}|{self.make}|{self.model}|({self.fuel_type})"


class ElectricVehicle(Vehicle):
    
    def __init__(self, make, model, year,battery_kwh, range_km ):
        print("Electric Vehicle is being created!")
        super().__init__( make, model, year, fuel_type = "Electric" )
        # new attributes
        self.battery_kwh    = battery_kwh
        self.range_km       = range_km
        self.range_km = 100   # starts fully 
        self.charge_percent = 100 

    def charge(self, percent):
        self.charge_percent = min(100, self.charge_percent + percent)
        print(f"{self.make} {self.model} charged to {self.charge_percent}%")

    def accelerate(self, amount): # method overriding
        self.speed += amount
        self.charge_percent -= 10
        print(f"{self.make} {self.model} accelerating to {self.speed} km/h")

    # method overriding --> form of polymorphism
    def __str__(self): ## how the object is serialised --> for when the object is to be shown in string or printed
        return f"{self.year} - {self.make} - {self.model} - ({self.fuel_type}) - {self.range_km} km"

    def __repr__(self): ## how the object is represented
        return f"{self.year}|{self.make}|{self.model}|({self.fuel_type}) | {self.range_km} km"  


class Tata_Nexon_EV(ElectricVehicle):
    """Specific model -- inherits from ElectricVehicle."""

    # Tata Nexon EV Long Range specs
    DEFAULT_BATTERY = 40.5   # kWh
    DEFAULT_RANGE   = 465    # km (claimed)

    def __init__(self, year, color):
        super().__init__(
            make="Tata", model="Nexon EV",
            year=year,
            battery_kwh=self.DEFAULT_BATTERY,
            range_km=self.DEFAULT_RANGE
        )
        self.color = color

    def __str__(self):
        return f"{super().__str__()} | Color: {self.color}"
```

```python
import datetime
class Employee:

    appraisal_pct = 0.1 ## class attribute
    organisation_name = "codeverra"
    
    def __init__(self, name, dob, deptt, salary):
        self.name = name # instance attribute
        self.deptt = deptt
        self.dob = dob
        # self.salary = salary # public attribute
        # self._salary = salary # protected attribute ( merely a convention )
        self.__salary = salary # private attribute ( merely a convention )
        # self.email = f"{self.name}@{Employee.organisation_name}.com"
        # self.age = self.calculate_age(self.dob)
        # self.age = 
  
    @property #--> this is a method but acting like a attribute
    def age(self):  # ramesh.age
        if self.authenticate():
             return self.calculate_age(self.dob)
        raise Exception("NOT AUTHORISED!") 
       
    @property
    def email(self):  
        '''
        this method will run whenever someone calls employee.email
        '''
        return f"{self.name}@{Employee.organisation_name}.com"
        
    @property  # getter method
    def salary(self): 
        if self.authenticate():
            return self.__salary
        raise Exception("NOT AUTHORISED!") 

    @salary.deleter
    def salary(self):
        print("this attribute is being deleted")
        
    # def email_update(func):

    #     '''
    #     '''

    # def log(func):
    #     '''
    #     '''
    
    ## enable the behaviour employee.salary = new_salary  ---> setter method
    # @email_update
    # @log
    @salary.setter
    def salary(self, new_salary):
        print("The salary is being changed!")
        self.__salary = new_salary
    
    def get_age(self):
        # self.age = self.calculate_age(self.dob)
        return self.calculate_age(self.dob)
    
    @staticmethod
    def calculate_age(dob):
        return datetime.datetime.now().year - datetime.datetime.strptime(dob, "%Y-%m-%d").year

    def authenticate(self):
        if random.randrange(0, 2) == 1:
            return True
        return False
    
    def show_salary(self):
        if self.authenticate():
            return self.__salary
        raise Exception("NOT AUTHORISED!")
        
    def give_appraisal(self): # instance method
        
        # self.salary = self.salary * (1 + Employee.appraisal_pct)
        self.__salary = self.__salary * (1 + self.appraisal_pct)
```
